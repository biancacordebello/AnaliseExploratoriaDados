```python
import pandas as pd
```

```python
arq = 'aluguel.csv'
dados = pd.read_csv(arq, sep=';')
dados
```

### Removendo Duplicadas

1 - Identificação da quantidadas de linhas duplicadas = 1175  
2 - Remoção das linhas duplicadas usando o "drop_duplicates()"


```python
duplicadas = dados[dados.duplicated()]
duplicadas
```

```python
dados = dados.drop_duplicates()
dados
```

#### Selecionar apenas as amostras do tipo: loft, flat, quitinete ou studio
