## 4. Selecionar apenas as amostras do tipo: **loft, flat, quitinete ou studio**

1. Filtragem do DataFrame por loft, flat, quitinete ou studio: `dados = dados.query('Tipo == "Loft" | Tipo == "Flat" | Tipo == "Quitinete" | Tipo == "Studio"')`
    
    Com essa filtragem será incluído apenas as linhas em que o valor da coluna 'Tipo' corresponde a 'Loft', 'Flat', 'Quitinete' ou 'Studio'. 

2. Exibição do DataFrame: `dados`
    Exibe a tabela, sendo possível ver que ela foi atualizada



```python
dados = dados.query('Tipo == "Loft" | Tipo == "Flat" | Tipo == "Quitinete" | Tipo == "Studio"')
dados
```

## 5. Preencher os valores nulos de IPTU e Condomínio com a moda dos valores

1. Cálculo da Moda em 'Condominio' e 'IPTU' utilizando o método usando o método `mode()` do pandas.

2. Converter valores para Inteiro: Utilizamos o método `iloc[0]` para acessar o primeiro valor da série de moda.

3. Preenchimento de Valores Nulos com as Modas utilizamos o método `fillna()` do pandas para preencher os valores nulos com valores específicos


```python
moda_condominio = dados['Condominio'].mode()
moda_iptu = dados['IPTU'].mode()

moda_condominio = int(moda_condominio.iloc[0])
moda_iptu = int(moda_iptu.iloc[0])

dados =  dados.fillna(value={'Condominio':moda_condominio,'IPTU':moda_iptu})
dados
```

## 6. Criar colunas custo_médio_mensal e custo_médio_mensal_por_m2 calculada a partir da seguinte fórmula:
**custo_medio_mensal = (Valor + Condomínio + IPTU/12)**


**custo_medio_mensal_por_m2 = custo_medio_mensal  / Área**


  - Para calcular o custo médio mensal utilizamos a fórmula e arredondamos esse valor para duas casas decimais.
  - Em seguida, calculamos o custo médio mensal por metro quadrado utilizando a mesma fórmula e também arredondando para duas casas decimais.
  - Os resultados são duas novas colunas no DataFrame `dados`, chamadas de `Custo Medio Mensal` e `Custo Medio Mensal por M2`.


```python
dados['Custo Medio Mensal'] = round(((dados['Valor'] + dados['Condominio'] + dados['IPTU']) / 12),2)
dados['Custo Medio Mensal por M2'] = round((dados['Custo Medio Mensal'] / dados['Area']),2)

dados
```

## 7. Criar um dataframe com o valor do custo médio mensal por m2 em cada um dos bairros.

   - Para calcular a média do custo médio mensal por metro quadrado para cada bairro, agrupamos o DataFrame `dados` pelos valores únicos na coluna 'Bairro' usando o `groupby()`.
   - Selecionamos a coluna 'Custo Medio Mensal por M2' e calculamos a média usando o método `mean()`.
   - O resultado é arredondado para duas casas decimais com `round()` e classificado em ordem decrescente de custo médio mensal por metro quadrado usando `sort_values()`.
   - O resultado final é atribuído à variável `media_bairro`.


```python
media_bairro = round(dados.groupby('Bairro')['Custo Medio Mensal por M2'].mean(),2).sort_values(ascending=False)
media_bairro
```

## 8. Criar um gráfico de barras com o custo médio mensal por m2 em cada bairro ordenado decrescentemente.
1. Visualização da Média do Custo Médio Mensal por Metro Quadrado por Bairro em um Gráfico de Barras Horizontais: Utilizando o método `plot()` do pandas com os argumentos `kind='barh'` para barras horizontais, `figsize=(7, 20)` para o tamanho da figura e `color='purple'` para a cor das barras.


```python
media_bairro.plot(kind='barh', figsize=(7, 20), color ='purple');
```
