```python
import pandas as pd
```

```python
arq = 'aluguel.csv'
dados = pd.read_csv(arq, sep=';')
dados
```

### Removendo Duplicadas

1 - Identificação da quantidadas de linhas duplicadas = 1175  
2 - Remoção das linhas duplicadas usando o "drop_duplicates()"


```python
duplicadas = dados[dados.duplicated()]
duplicadas
```

```python
dados = dados.drop_duplicates()
dados
```

### REMOVER NULOS

1 - Identificação da quantidade de valores nulos em cada coluna  
      * Tipo = 16  
    * Bairro = 0  
    - Quartos = 0  
    - Vagas = 0  
    - Suites = 0  
    - Area = 0  
    - Valor = 17  
    - Condominio = 3949  
    - IPTU = 9887  

2 - 

```python
dados.isnull().sum()
```

```python
dados = dados.dropna(subset=['Valor'])
dados.isnull().sum()
```

#### Selecionar apenas as amostras do tipo: loft, flat, quitinete ou studio



```python
dados = dados.query('Tipo == "Loft" | Tipo == "Flat" | Tipo == "Quitinete" | Tipo == "Studio"')
dados
```

```python
moda_condominio = dados['Condominio'].mode()
moda_iptu = dados['IPTU'].mode()

moda_condominio = int(moda_condominio.iloc[0])
moda_iptu = int(moda_iptu.iloc[0])

dados =  dados.fillna(value={'Condominio':moda_condominio,'IPTU':moda_iptu})
dados
```

```python
dados['Custo Medio Mensal'] = round(((dados['Valor'] + dados['Condominio'] + dados['IPTU']) / 12),2)
dados['Custo Medio Mensal por M2'] = round((dados['Custo Medio Mensal'] / dados['Area']),2)

dados
```

```python
media_bairro = round(dados.groupby('Bairro')[['Custo Medio Mensal por M2']].mean(),2).sort_values('Custo Medio Mensal por M2', ascending=False)
media_bairro
```

```python
media_bairro.plot(kind='barh', figsize=(7, 20), color ='skyblue');
```

```python
custo_menor_mil = dados[dados['Custo Medio Mensal'] < 1000]
print(f"Há cerca de {len(custo_menor_mil)} imóveis abaixo de 1000")

dataset = custo_menor_mil.groupby('Bairro')
bairros = dataset.size()

print(f"Estão localizados nos bairros: \n{bairros}")
```
